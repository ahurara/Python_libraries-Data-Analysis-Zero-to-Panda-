# Inoder to convert date ro object type to date time type

```python
covid_df["dates"]=pd.to_datetime(covid_df.date)
covid_df['dates']
```

```python
covid_df["year"]=pd.DatetimeIndex(covid_df.dates).year
covid_df['month']=pd.DatetimeIndex(covid_df.dates).month
covid_df['day']=pd.DatetimeIndex(covid_df.dates).day
covid_df['weekday']=pd.DatetimeIndex(covid_df.dates).weekday
covid_df
```

```python
covid_df_may=covid_df[covid_df.month ==5]
covid_df_may
```

```python
covid_df_may_metrics=covid_df_may[['new_cases','new_deaths','new_tests']]
covid_may_total=covid_df_may_metrics.sum()
covid_may_total
```

```python
covid_sunday_mean=covid_df[covid_df.weekday==6]
covid_sunday_mean.new_cases.mean()
```

```python
covid_month_df=covid_df.groupby('month')[['new_cases','new_tests','new_deaths']].sum()
covid_month_df
```
